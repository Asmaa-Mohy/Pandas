##### Maps
It is possible to iterate over a DataFrame or Series as you would with a list, but doing so — especially on large datasets — is very slow.

map is function that takes one set of values and "maps" them to another set of values.
We can transform data from the format it is in now to the format that we want it using Maps.

There are two mapping methods:
- `map()` method only works on a pandas series,The function you pass to `map()` should expect a single value from the Series and return a transformed version of that value
- `apply` method can be applied both to series and dataframes,if we want to transform a whole DataFrame by calling a custom method on each row.

###### NOTE:
`map()` and `apply` don't modify the original data they're called on




```python
def str_to_int(strk):
    return float(strk[:-1]) if 'K'in strk else float(strk)/1000.0
    
str_to_int('2.8K')
```

```python
data_f.loc[:,'claps'].apply(str_to_int)
```

```python
data_f.loc[:,'claps'].map(str_to_int)
```

```python
def str_to_int2(row):
    return float(row.claps[:-1]) if 'K'in row.claps else float(row.claps)/1000.0


data_f.apply(str_to_int, axis=1)
#data_f.map(str_to_int, axis=1) Error
```

###### correlation
we use correlation to discover relationships between variables, the value of correlation coefficient is between 1,-1 
- correlation coefficient increase (close to 1) , means that there is linear relation(directly proportionality or  positive increase).
- correlation coefficient decrease (close to -1) , means that there is linear relation( inversely proportionality or  negative decrease).
- correlation coefficient is close to 0 , means that there isn’t a positive or negative increase. 

```python
data_f.corr(method ='pearson')
```

###### Dealing Data Missing
Entries missing values are given the value NaN.

To select NaN entries we can use `isnull()` .

To remove all columns or rows that contain least one missing value we can use `dropna()` .

To fill in missing values in a dataframe OR specify what we want the NaN values to be replaced with we can use `fillna()` .

```python
article_data.isnull().sum()
```

```python
#article_data.dropna() #remove all the rows that contain a missing value
article_data.dropna(axis=1) #remove all columns with at least one missing value
```

```python
article_data.loc[:,['id','url','responses','publication','date']]
```

```python
article_data.loc[:,['id','url','responses','publication','date']].fillna("Unknown")
```

###### Parsing Dates
is convert dates it so that it is recognized as a date. 

The basic idea is that you need to point out which parts of the date are where and what punctuation is between them. but the most common are :
- `%d` for day,
- `%m` for month,
- `%y` for a two-digit year 
- `%Y` for a four digit year

For example:

   - 1/17/07 has the format `%m/%d/%y`



```python
article_data.loc[:, 'date']
```

```python
article_data.loc[:, 'date'].fillna('0000-00-00') 
# check date we find that the dtype of date "object" it will have strings in it.
```

```python
article_data['date_parsed']=pd.to_datetime(article_data['date'], format="%Y-%m-%d")
article_data['date_parsed'][-100:]
```

```python
#Select the day of the month
day_of_month=article_data['date_parsed'].dt.day
day_of_month
```
